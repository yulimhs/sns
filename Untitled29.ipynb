```python
import os
import pandas as pd
from selenium import webdriver
from bs4 import BeautifulSoup
import time
import random
import math
import numpy
import re
import sys


print("-"*80)
print("연습문제 8-1 네이버 영화 리뷰 정보 수집하기")
print("="*80)
query_txt = input('1. 크롤링 할 영화의 제목을 입력하세요:')
cnt = int(input('2. 크롤링 할 리뷰건수는 몇건입니까?:'))
page_cnt = math.ceil(cnt/10)

f_dir = input('3. 파일을 저장할 폴더명만 쓰세요(예:c:\py_temp\):')

#1. 저장 폴더 파일명 : 시간부터 
now = time.localtime()
s = "%04d-%02d-%02d-%02d-%02d-%02d" %(now.tm_year, now.tm_mon, now.tm_mday, now.tm_hour, now.tm_min, now.tm_sec)

os.makedirs(f_dir+s+'-'+query_txt)
os.chdir(f_dir+s+'-'+query_txt)

ff_name = f_dir+s+'-'+query_txt+'\\'+s+'-'+query_txt+'.txt'
fc_name = f_dir+s+'-'+query_txt+'\\'+s+'-'+query_txt+'.csv'
fx_name = f_dir+s+'-'+query_txt+'\\'+s+'-'+query_txt+'.xls'

#2. s 타임하고 크롬 드라이버 경로

s_time = time.time()

path = "c:/tmp/chromedriver_win32/chromedriver.exe"
driver = webdriver.Chrome(path)
driver.get("https://movie.naver.com")
time.sleep(2)

#3. 제목검색

driver.find_element_by_id("ipt_tx_srch").click()
element = driver.find_element_by_id("ipt_tx_srch")
element.send_keys(query_txt)
driver.find_element_by_class_name("btn_srch").click() #검색버튼
driver.find_element_by_xpath('''//*[@id="old_content"]/ul[2]/li/dl/dt/a''').click() #일치결과

driver.find_element_by_xpath('''//*[@id="movieEndTabMenu"]/li[5]/a''').click() #평점

#4. 리뷰 페이지
driver.switch_to.frame('pointAfterListIframe') #attribute error 해결 find 
html = driver.page_source # 체크
soup = BeautifulSoup(html,'html.parser')

# 총 카운트랑 페이지당 10건
result= soup.find('div', class_='score_total').find('strong','total').find('em').get_text()

print("=" *80)
result2 = result.replace(",","")
search_cnt = int(result2)
print(search_cnt) # 총 카운트 확인

if cnt > search_cnt:
    cnt = search_cnt

print('전체 검색 결과: %s' %search_cnt)
print('출력 건수: %s' %cnt)
print('크롤링 페이지 수: %s' %page_cnt)
print("=" *80)

score2=[]
review2=[]
writer2=[]
date2=[]
good2=[]
bad2=[]

count = 0
now_page = 1 # 페이지번호

while (now_page <= page_cnt) :
    html = driver.page_source
    soup = BeautifulSoup(html,'html.parser')
    
    list = soup.find('div', class_="score_result").find('ul').find_all('li')
    
    for li in list:
        count += 1
        f = open(ff_name,'a',encoding="utf-8")
        print("총 %s 건 중 %s 번째 리뷰 데이터를 수집합니다===" %(cnt,count))
        score = li.find('div',class_='star_score').find('em').get_text()
        print('1.별점: %s ' %score)
        f.write("\n")
        f.write("총 %s 건 중 %s 번째 리뷰 데이터를 수집합니다===" %(cnt,count)+"\n")
        f.write("1.별점:"+score+"\n")
        score2.append(score)
        
        review = li.find('div',class_='score_reple').find('p').get_text()
        review = review.strip() # 관람객일시 strip 이 안됨..
        print('2.리뷰내용: %s ' %review)
        f.write('2.리뷰내용:'+review+"\n")
        review2.append(review)
        
        wb = li.find('div',class_='score_reple').find_all('em') # 한줄에 두개라서
        writer = wb[0].find('span').get_text()
        print("3.작성자: %s " %writer)
        writer = writer.strip()
        f.write("3.작성자:"+writer+"\n")
        writer2.append(writer)
        
        date = wb[1].get_text()
        print('4.작성일자: %s' %date)
        f.write("4.작성일자:"+date+"\n")
        date2.append(date)
        
        gb = li.find('div',class_='btn_area').find_all('a') # 한줄에 두개라서
        good = gb[0].find('strong').get_text()
        print("5.공감:%s" %good)
        f.write("5.공감:"+good+"\n")
        good2.append(good)
        
        bad = gb[1].find('strong').get_text()
        print("5.비공감:%s" %bad)
        f.write("5.비공감:"+bad+"\n")
        bad2.append(bad)        
        
        if count == cnt:
            break
        
    now_page += 1
    
    if now_page > page_cnt:
        break
    else:
        driver.find_element_by_link_text("%s" %now_page).click()

        
# xls, csv 저장
movie_review = pd.DataFrame()
movie_review['별점(평점)']=score2
movie_review['리뷰내용']=review2
movie_review['작성자']=writer2
movie_review['작성일자']=date2
movie_review['공감횟수']=good2
movie_review['비공감횟수']=bad2

movie_review.to_csv(fc_name, encoding='utf-8-sig', index=True)
movie_review.to_excel(fx_name, index=True)

e_time = time.time()
t_time = e_time - s_time

print("\n") 
print("=" *80)
print("1.요청된 총 %s 건의 리뷰 중에서 실제 크롤링 된 리뷰수는 %s 건입니다" %(cnt,count))
print("2.총 소요시간은 %s 초 입니다 " %round(t_time,1))
print("3.파일 저장 완료: txt 파일명:%s" %ff_name)
print("4.파일 저장 완료: csv 파일명:%s" %fc_name)
print("5.파일 저장 완료: xls 파일명:%s" %fx_name)
print("=" *80)

driver.close( )
```
